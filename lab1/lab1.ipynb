```python
from matplotlib import image
import matplotlib.pyplot as plt
import numpy as np

import os

IMAGES_DIR_PATH = os.getcwd() + "/images/"
K = [1, .8, .5, .2]
L = [0, 20, 50, 80]
```

```python
def linear_1d_filtering(fname, k=1., l=0) -> np.ndarray:
    img = image.imread(fname=IMAGES_DIR_PATH + fname)
    img = img.astype(float)
    img = img*k + l
    img[img > 255] = 255
    img[img < 0] = 0
    return img.astype(int)
```

```python
def filtering_visualization(fname):
    fig, axs = plt.subplots(4, 4, figsize=(36,20))
    for i in range(4):
        for j in range(4):
            axs[i, j].set_axis_off()
            axs[i, j].set_title(f"{fname.rsplit('.')[0]} : K = {K[i]}, L = {L[j]}", fontsize=24)
            axs[i, j].imshow(linear_1d_filtering(fname, K[i], L[j]))
```

```python
for fname in os.listdir(IMAGES_DIR_PATH):
    filtering_visualization(fname)
```
